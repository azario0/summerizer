# Analyse de valeurs unique 

```python
# Obtenez les types de données de toutes les colonnes
data_types = df.dtypes

# Filtrer pour les types de données 'object'
object_dtypes = data_types[data_types == 'object']

# Obtenez les noms des colonnes 'object' sous forme de liste
object_columns = object_dtypes.index.tolist()

# Afficher uniquement les colonnes 'object'
object_df = df[object_columns]
object_df

# Parcourir les colonnes 'object' et obtenir les valeurs uniques
for column in object_columns:
    unique_values = object_df[column].unique()
    num_unique = len(unique_values)
    print(f"Valeurs uniques pour la colonne '{column}' (Total: {num_unique}):")

    # Afficher toutes les valeurs uniques si elles ne sont pas trop nombreuses, sinon seulement les premières 10
    if num_unique <= 5:  # Ajustez ce seuil selon les besoins
        print(unique_values)
    else:
        print(f"Affichage des premières 10 sur {num_unique} valeurs uniques :")
        print(unique_values[:5]) # Afficher les premières 10
    print("-" * 30)  # Séparateur pour plus de clarté

null_counts = df.isnull().sum()
# Filtrer pour les colonnes avec au moins une valeur nulle
columns_with_nulls = null_counts[null_counts > 0]
if columns_with_nulls.empty :
    print('Aucune valeur nulle dans le jeu de données')
else :
    print(f"Colonnes avec des nulls : \n{columns_with_nulls}")

```

# Describe du dataset


```python
def round_to_0_02(x):
    """Arrondit un nombre au multiple le plus proche de 0,02."""
    if pd.isna(x):  # Gérer les valeurs NaN
        return "NaN"
    return round(x / 0.02) * 0.02

def format_description(df, filename="description.txt"):
    """Génère et écrit une description formatée dans un fichier, en arrondissant les nombres à 0,02."""

    description = df.describe()

    # Appliquer l'arrondi à toutes les colonnes numériques de la description
    for col in description.columns:
        if pd.api.types.is_numeric_dtype(description[col]): # vérifier si numérique
          description[col] = description[col].apply(round_to_0_02)


    return description



format_description(df)
```
